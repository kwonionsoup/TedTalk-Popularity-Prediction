```python
# Import necessary libraries
import numpy as np
import pandas as pd
import ast
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
# Read in file and view data
df = pd.read_csv("data\date_popularity_processed.csv")
df[0:5]
```

```python
# View summary of columns
df.describe()
```

```python
# scatter plot of likes vs views
plt.figure(figsize=(10, 6))
plt.scatter(df['views'], df['likes'], marker='o')
plt.title('Popularity of TED Talks by Video ID')
plt.xlabel('Views')
plt.ylabel('Likes')
plt.grid(True)
plt.show()
```

```python
# scatter plot of popularity
plt.figure(figsize=(10, 6))
plt.scatter(df['_id'], df['popularity'], marker='o')
plt.title('Popularity of TED Talks by Video ID')
plt.xlabel('Video ID')
plt.ylabel('Popularity')
plt.grid(True)
plt.show()
```

```python
df['recorded_date'] = pd.to_datetime(df['recorded_date'])
talks_per_year = df.groupby(df['recorded_date'].dt.year).size()
talks_per_year.plot(kind='line')
plt.xlabel('Year')
plt.ylabel('Number of Talks')
plt.title('Number of TED Talks Recorded per Year')

# Show the plot
plt.show()
```

```python
df['published_date'] = pd.to_datetime(df['published_date'])
talks_per_year = df.groupby(df['published_date'].dt.year).size()
talks_per_year.plot(kind='line')
plt.xlabel('Year')
plt.ylabel('Number of Talks')
plt.title('Number of TED Talks Published per Year')

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['recorded_date'], df['popularity'], marker='o')
plt.title('Popularity of TED Talks Over Time')
plt.xlabel('Date Recorded')
plt.ylabel('Popularity')
plt.grid(True)
plt.show()
```
