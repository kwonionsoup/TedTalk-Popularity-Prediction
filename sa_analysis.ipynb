```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("data/data_sentiment_analysis.csv")
data[0:5]
```

```python
data.describe()
```

```python
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
plt.scatter(data['_id'], data['tb_score'], label='TextBlob', marker='o')
plt.scatter(data['_id'], data['vd_score'], label='Vader', marker='o')

plt.xlabel('ID')
plt.ylabel('Sentiment Analysis Score')
plt.title('Comparison of TextBlob and Vader')
plt.legend()  # Show legend

plt.grid(True)  # Add grid
plt.show()
```

```python
popularity_data = pd.read_csv("data/date_popularity_processed.csv")
popularity_data[0:5]
```

```python
data['popularity'] = popularity_data['popularity']
data.describe()
```

```python
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
plt.scatter(data['_id'], data['tb_score'], label='TextBlob', marker='o')
plt.scatter(data['_id'], data['vd_score'], label='Vader', marker='o')
plt.scatter(data['_id'], data['popularity'], label='Popularity', marker='o')

plt.xlabel('ID')
plt.ylabel('Sentiment Analysis Score')
plt.title('Comparison of TextBlob, Vader, and Popularity (Standardized)')
plt.legend()  # Show legend

plt.grid(True)  # Add grid
plt.show()
```

```python
# Standardize popularity between -1 and 1
min_val = data['popularity'].min()
max_val = data['popularity'].max()
data['popularity_standardized'] = 2 * ((data['popularity'] - min_val) / (max_val - min_val)) - 1
data.head()
```

```python
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
plt.scatter(data['_id'], data['tb_score'], label='TextBlob', marker='o')
plt.scatter(data['_id'], data['vd_score'], label='Vader', marker='o')
plt.scatter(data['_id'], data['popularity_standardized'], label='Popularity', marker='o')

plt.xlabel('ID')
plt.ylabel('Sentiment Analysis Score')
plt.title('Comparison of TextBlob, Vader, and Popularity (Standardized)')
plt.legend()  # Show legend

plt.grid(True)  # Add grid
plt.show()
```

```python
data['published_date'] = pd.to_datetime(data['published_date'])

plt.figure(figsize=(10, 6))
plt.scatter(data['published_date'], data['tb_score'], marker='o')
plt.title('TextBlob Score of TED Talks Over Time')
plt.xlabel('Date Published')
plt.ylabel('TextBlob Score')
plt.grid(True)
plt.show()
```
