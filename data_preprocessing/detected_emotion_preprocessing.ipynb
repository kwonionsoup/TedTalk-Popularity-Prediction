```python
import pandas as pd

# import 'updated_trascripts_with_emotion_v2.csv' 
# import 'updated_transcripts_with_emotion_v2.csv'  
data = pd.read_csv('/Users/taeeunkwon/Downloads/vscode_projects/data/updated_transcripts_with_emotions_v2.csv')
data
```

```python
def parse_json(data):
    import json
    try:
        # Load JSON content if it's a string, else return an empty dictionary
        return json.loads(data) if isinstance(data, str) else {}
    except json.JSONDecodeError:
        # Return an empty dictionary in case of JSON decode error
        return {}


data['detected_emotion'] = data['detected_emotion'].apply(parse_json)
```

```python
emotion_df = pd.json_normalize(data['detected_emotion']).fillna(0)
data = pd.concat([data.drop(['detected_emotion'], axis=1), emotion_df], axis=1)

```
